```python
def selection_sort(num):
    n= len(num)
    for i in range (0,n):
        min_ind=i
        for j in range(i+1,n):
            if num[j] < num[min_ind]:
                min_ind = j
        num[i],num[min_ind]  =num[min_ind],num[i]     
        
    return num
num =[9,3,7,8,2,1,6,5,4]
print (selection_sort(num))
```
